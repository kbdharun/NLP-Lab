```python
from collections import Counter
from nltk.tokenize import TreebankWordTokenizer

# Read the text file
with open("nlp.txt") as f:
    text = f.read()

# Tokenize the text
tokenizer = TreebankWordTokenizer()
tokens = tokenizer.tokenize(text)

# Display total number of unique words
unique_words = set(tokens)
total_unique_words = len(unique_words)
print("Total number of unique words:", total_unique_words)

# Frequency of each word without duplicate entries
word_frequency = Counter(tokens)
print("\nWord frequencies:")
for word, freq in word_frequency.items():
    print(word, freq)

# Joint probability of each word
joint_probabilities = Counter(zip(tokens, tokens[1:]))
total_pairs = len(joint_probabilities)
print("\nTotal number of unique pairs:", total_pairs)
print("Joint probabilities:")
for pair, freq in joint_probabilities.items():
    print(" ".join(pair), freq)

# The pair with the maximum frequency
max_freq_pair = max(joint_probabilities, key=joint_probabilities.get)
print("\nThe pair with the max frequency:", " ".join(max_freq_pair))
 
# Apply part-of-speech tagging to the tokens
tagged = nltk.pos_tag(tokens)
 
# Apply named entity recognition to the tagged words
entities = nltk.chunk.ne_chunk(tagged)

# Print the entities found in the text
for entity in entities:
    if hasattr(entity, 'label') and entity.label() == 'ORGANIZATION':
        print(entity.label(),'-->', ''.join(c[0] for c in entity))
    elif hasattr(entity, 'label') and entity.label() == 'GPE':
        print(entity.label(), '-->',''.join(c[0] for c in entity))
    elif hasattr(entity, 'label') and entity.label() == 'PERSON':
        print(entity.label(), '-->',''.join(c[0] for c in entity))
```

```python
import re
import nltk
from nltk.tokenize import TreebankWordTokenizer
from nltk.stem import PorterStemmer
f=open("nlp.txt",'r')
tk=TreebankWordTokenizer()
new_string = re.sub(r'[^\w\s]', '', f.read())
val=tk.tokenize(new_string)
tagged=nltk.pos_tag(val)
entities=nltk.chunk.ne_chunk(tagged)

for entity in entities:
    if hasattr(entity, 'label') and entity.label() == 'ORGANIZATION':
        print(entity.label(),'-->', ''.join(c[0] for c in entity))
    elif hasattr(entity, 'label') and entity.label() == 'GPE':
        print(entity.label(), '-->',''.join(c[0] for c in entity))
    elif hasattr(entity, 'label') and entity.label() == 'PERSON':
        print(entity.label(), '-->',''.join(c[0] for c in entity))
f.close()
```
