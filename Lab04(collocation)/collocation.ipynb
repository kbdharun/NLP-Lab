```python
import nltk
from nltk.tokenize import word_tokenize
import string

#setting string
text = 'Coffee has been forever been close to my heart. If you ask me, I don’t really remember when exactly I fell in love, but I just did. The aroma, the taste, the flavour, everything is so therapeutic. I am sure, like me, you have wondered what is about coffee that we love so much? Like, how is it even possible to love coffee so much?My love for coffee extends far back, atleast a decade back when I used to sip some coffee here and there. It then grew into a weekly Saturday morning tradition almost, when my friends would either come over to my place or we would all go out, meet at a coffee shop and nonchalantly order for the beverage of our choice – coffee. My love for coffee has developed over the years, but has grown more than ever, especially, in the past few years. I know it may seem kind of silly to some to love something like a beverage as much as I do, but then again, it’s more than just a “simple beverage” to me.You see, now that I think about it, I feel, there is no smell I love waking up to than a fresh brewed cup of coffee. I think, there is just something so inviting about the smell of coffee that makes me feel so ready; I mean, it’s almost like it makes me feel renewed. I actually love the aroma of coffee so much that I even have coffee scented perfumes that I use! Of course, there is the taste that I truly adore, it was never something that I had to acquire, so to speak. I appreciate the taste of coffee without any cream or sugar, quite frankly.I don’t know why am I getting into the tiny details about its aroma or taste in this piece of blog post, but I totally feel, I just get lost when it comes to either talking about coffee or penning down about the subject.At the end of it all, I would want to say one simple thing – when you appreciate coffee for more than the caffeine it provides, that’s when you truly begin to love coffee. That’s when you realize, you have a heart of coffee…'

#preparing the string
translating = str.maketrans('', '', string.punctuation)
cleaned_text = text.translate(translating)
token = word_tokenize(cleaned_text.lower())

#printing the token
print(token)
```

```python
#FINDING THE FREQ NOW OF PAIRS OF WORDS
Elements_count = {}
for i in range(len(token)):
    if(i != len(token)-1):
        a = token[i]+token[i+1]
        if a in Elements_count:
            Elements_count[a] += 1
        else:
            Elements_count[a] = 1
            
#SORTING THE DICTIONARY
sorted_dict = dict(sorted(Elements_count .items(), key=lambda item: item[1]))

#PRINTING THE DICTIONARY
for key, value in sorted_dict.items():
   print(f"{key}->{value}",end='  ')
```

```python
#NOW FINDING HOW MANY TIMES 'i' HAS COME
i_count = 0
for j in range(len(token)):
    if token[j] == 'i':
        i_count +=1
print("The number of times \'i\' has come:",i_count)
```

```python
#SIMILARLY FINDING FOR 'that'
that_count = 0
for j in range(len(token)):
    if token[j] == 'that':
        that_count +=1
print("The number of times \'that\' has come:",that_count)
```

```python
#TOTAL NUMBER OF WORDS
total_count = len(token)
print("Total count of toekn :",total_count)
```

```python
import math

#T-TEST FOR 'thati'
Null_hypothesis = (i_count/total_count)*(that_count/total_count)
print("The null hypothesis is :",Null_hypothesis)
observed = (5/total_count)
print("The observed is :", observed)

t_test = (observed-Null_hypothesis)/(math.sqrt(math.pow(observed,2)/total_count))
print("The value of t-test is :",t_test)

print("Since the confidence level id 0.005 and ",t_test,'>',"3.169 , we can reject it and Hence it is a collocation")
```

```python
#NOW HYPOTHESIS TEST
thati_count = 0
that_other_count = 0
other_i_count = 0
other_count = 0

for j in range(len(token)):
    if(j< len(token)-1):
        if(token[j] == 'that' and token[j+1] == 'i'):
            thati_count +=1
        elif(token[j] == 'that' and token[j+1] != 'i'):
            that_other_count += 1
        elif(token[j] != 'that' and token[j+1] == 'i'):
            other_i_count += 1
        elif(token[j] != 'that' and token[j+1] != 'i'):
            other_count += 1
            
#PRINTING
print(thati_count)
print(that_other_count) 
print(other_i_count)
print(other_count)
```

```python
#PRINTING
print("The value of that i :",(i_count*that_count)/total_count)
print("The value of that other :",(that_count*that_other_count)/total_count)
print("The value of other i :",(i_count*other_i_count)/total_count)
print("The value of other :",(other_count*total_count)/total_count)
```

```python
i =(i_count*that_count)/total_count
j = (that_count*that_other_count)/total_count
k =(i_count*other_i_count)/total_count
l = (other_count*total_count)/total_count
```

```python
#NOW x^2
a = math.pow(thati_count-i,2)/i
b = math.pow(that_other_count-j,2)/j
c = math.pow(other_i_count-k,2)/k
d = math.pow(other_count-l,2)/l
hypothesis =a+b+c+d
print("The hypothesis test is :",hypothesis)
```
