```python
from collections import Counter
from nltk.tokenize import TreebankWordTokenizer

# Read the text file
with open("test.txt") as f:
    text = f.read()

# Tokenize the text
tokenizer = TreebankWordTokenizer()
tokens = tokenizer.tokenize(text)

# Task 1: Display total number of unique words
unique_words = set(tokens)
total_unique_words = len(unique_words)
print("Total number of unique words:", total_unique_words)

# Task 2: Frequency of each word without duplicate entries
word_frequency = Counter(tokens)
print("\nWord frequencies:")
for word, freq in word_frequency.items():
    print(word, freq)

# Task 3: Joint probability of each word
joint_probabilities = Counter(zip(tokens, tokens[1:]))
total_pairs = len(joint_probabilities)
print("\nTotal number of unique pairs:", total_pairs)
print("Joint probabilities:")
for pair, freq in joint_probabilities.items():
    print(" ".join(pair), freq)

# The pair with the maximum frequency
max_freq_pair = max(joint_probabilities, key=joint_probabilities.get)
print("\nThe pair with the max frequency:", " ".join(max_freq_pair))
```
