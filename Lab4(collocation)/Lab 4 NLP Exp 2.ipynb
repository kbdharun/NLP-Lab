```python
R_patterns = [
   (r'won\'t', 'will not'),
   (r'can\'t', 'cannot'),
   (r'i\'m', 'i am'),
   (r'(\w+)\'ll', '\g<1> will'),
   (r'(\w+)n\'t', '\g<1> not'),
   (r'(\w+)\'ve', '\g<1> have'),
   (r'(\w+)\'s', '\g<1> is'),
   (r'(\w+)\'re', '\g<1> are'),
]
```

```python
class REReplacer(object):
   def __init__(self, pattern = R_patterns):
      self.pattern = [(re.compile(regex), repl) for (regex, repl) in pattern]
   def replace(self, text):
      s = text
      for (pattern, repl) in self.pattern:
         s = re.sub(pattern, repl, s)
      return s
```

```python
rep_word = REReplacer()
text = rep_word.replace(text)
```

```python
text = text.replace(',','')
text = text.replace('.','')
text = text.replace('\'','')
text = text.replace('"','')
text = text.replace('’','')
text = text.replace('\n','')
text = text.lower()
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer_wrd = TreebankWordTokenizer()
tokenized = tokenizer_wrd.tokenize(text)
tokenized
```

```python
d = {}
for word in tokenized:
    if word not in d:
        d[word] = 1
    else:
        d[word] += 1
```

```python
nuws = len(d)
nws = len(tokenized)
print("The number of words : ",str(nws))
print("The number of unique words : ",str(nuws))
print("Frequencies of the words.....")
print(" ")

#for i in d:
#    print(i," ",str(d[i]))
```

```python
#Bigram
jd = {}
for i in range(len(tokenized)):
    w1 = i
    w2 = i+1
    if(w2 != len(tokenized)):
        count = 0
        wd1 = tokenized[w1]
        wd2 = tokenized[w2]
        for j in range(len(tokenized)):
            if(tokenized[j] == wd1 and tokenized[j+1] == wd2):
                count += 1
        jd[str(wd1 + " " + wd2)] = count
'''for i in jd:
    print("Joint probability of ",i," : ",str(jd[i]/nws))'''
    
maxp = 0
maxq = 0
maxr = 0
for i in jd:
    if jd[i] > maxp:
        maxp = jd[i]
    elif jd[i] >maxq and jd[i]<= maxp:
        maxq = jd[i]
    elif jd[i] > maxr and jd[i] <= maxr and jd[i] <= maxp:
        maxr = jd[i]
print("Maximum probability bigram pair ....")
j = 0
top_3 = []
for i in jd:
    if jd[i] == maxp:
        print(i,' ',str(jd[i]))
        top_3.append(i)
        j += 1
        if(j>=3):
            break
    elif jd[i] == maxq:
        print(i,' ',str(jd[i]))
        top_3.append(i)
        j += 1
        if(j>=3):
            break
    elif jd[i] == maxr:
        print(i,' ',str(jd[i]))
        top_3.append(i)
        j += 1
        if(j>=3):
            break
```

# t - test

```python
for i in top_3:
    wrd1 = i.split(' ')[0]
    wrd2 = i.split(' ')[1]
    p_wrd1 = float(d[wrd1]/nws)
    p_wrd2 = float(d[wrd2]/nws)
    mu = float(p_wrd1*p_wrd2)
    mean = float(jd[i]/nws)
    s_sq = float(mean*float(1-mean))
    t_numerator = float(mean - mu)
    t_denominator = float((s_sq/nws)**(0.5))
    t_score = float(t_numerator/t_denominator)
    print("The t score for '"+i+"' is : "+str(t_score))
    
```

# Pearson's chi-square test

```python
for i in top_3:
    wrd1 = i.split(' ')[0]
    wrd2 = i.split(' ')[1]
    c_wrd1 = d[wrd1]
    c_n_wrd1 = nws - d[wrd1]
    c_wrd2 = d[wrd2]
    c_n_wrd2 = nws - d[wrd2]
    o_w1w2 = jd[i]
    o_nw1w2 = c_wrd2 - o_w1w2
    o_w1nw2 = c_wrd1 - o_w1w2
    o_nw1nw2 = c_n_wrd1 - o_nw1w2
    e_w1w2 = float((c_wrd1*c_wrd2)/nws)
    e_nw1w2 = float((c_n_wrd1*c_wrd2)/nws)
    e_w1nw2 = float((c_n_wrd2*c_wrd1)/nws)
    e_nw1nw2 = float((c_n_wrd2*c_n_wrd1)/nws)
    x_w1w2 = float(((o_w1w2 - e_w1w2)**2)/e_w1w2)
    x_nw1w2 = float(((o_nw1w2 - e_nw1w2)**2)/e_nw1w2)
    x_w1nw2 = float(((o_w1nw2 - e_w1nw2)**2)/e_w1nw2)
    x_nw1nw2 = float(((o_nw1nw2 - e_nw1nw2)**2)/e_nw1nw2)
    print("Observed : ",o_w1w2,o_nw1w2,o_w1nw2,o_nw1nw2)
    #print("Observed Total : ",o_w1w2+o_nw1w2+o_w1nw2+o_nw1nw2)
    print("Expected : ",e_w1w2,e_nw1w2,e_w1nw2,e_nw1nw2)
    #print("Expected Total : ",e_w1w2+e_nw1w2+e_w1nw2+e_nw1nw2)
    chi_sq = float(x_w1w2 + x_nw1w2 + x_w1nw2 + x_nw1nw2)
    print("X^2      : ",x_w1w2 , x_nw1w2 , x_w1nw2 , x_nw1nw2)
    print("The chi-square score for '"+i+"' is : "+str(chi_sq))
```
